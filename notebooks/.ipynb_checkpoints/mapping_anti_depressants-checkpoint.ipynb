```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
import numpy as np
import json
from tqdm import tqdm
sns.set(rc={'figure.figsize':(12,8)})
%matplotlib inline
```

```python
presc = pd.read_csv('../data_prep/depression_V4.csv.gz', compression='gzip')
wards = pd.read_csv('../data_prep/Wards.csv')
```

```python
wards.head(n=4)
```

```python
presc.head(n=4)
```

```python
year = []
value = []
labels = []
cost = []
month = 1
for name , group in presc.groupby('YYYYMM'):
    year.append(month)
    labels.append(name)
    value.append(sum(group['Total_items'])/sum(group['Patient_count']))
    cost.append(sum(group['Total_cost'])/sum(group['Patient_count']))
    month+=1
    
fig, ax = plt.subplots()
fig.set_size_inches(16, 8)
sns.set_style("white")
plt.grid(False)
plt.plot(year,value)
plt.plot(year,cost)
plt.legend(['Prescription items per capita', 'Prescription cost per capita (GBP)'])
plt.xticks(year,labels,rotation = 90)
```

```python
import geopandas as gpd
map_london = gpd.read_file('/datasets/sagarj/geofiles/London-wards-2014 (1)/London-wards-2014_ESRI/London_Ward.shp')
```

```python
map_london.head(n=2)
```

```python
map_london.plot()
```

```python
presc_2019 = presc[(presc['YYYYMM']>=201901) & (presc['YYYYMM'] <= 201912) ]
```

```python
presc_2019_with_wards = pd.merge(presc_2019,wards,left_on='LSOA_CODE' , right_on='LSOA11CD')
```

```python
presc_2019_with_wards.head(n=4)
```

```python
ward_level_dict = {'Ward_code' : [] , 'items_per_capita' : [] , 'cost_per_capita' : []}
for name , group in tqdm(presc_2019_with_wards.groupby('WD15CD')):
    ward_level_dict['Ward_code'].append(name)
    ward_level_dict['items_per_capita'].append(sum(group['Total_items'])/(sum(group['All Ages'])/12))
    ward_level_dict['cost_per_capita'].append(sum(group['Total_cost'])/(sum(group['All Ages'])/12))
ward_level_df = pd.DataFrame.from_dict(ward_level_dict)
```

```python
ward_level_df.head(n=4)
```

```python
london_wards_df = pd.merge(map_london,ward_level_df,left_on='GSS_CODE',right_on='Ward_code')
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20,20)
ax.axis('off')
ax.set_title('Anti-depressant prescriptions per capita', fontdict={'fontsize': '25', 'fontweight' : '500'})
sm = plt.cm.ScalarMappable(cmap='GnBu', norm=plt.Normalize(vmin=london_wards_df['items_per_capita'].min(), vmax=london_wards_df['items_per_capita'].max()))
#empty array for the data range
sm._A = []
#add the colorbar to the figure
cbar = fig.colorbar(sm,shrink=0.3)
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(25)
london_wards_df.plot(ax=ax,column='items_per_capita', cmap='GnBu', linewidth=0.8)
```

```python
lad_boundries = gpd.read_file('../data_prep/geography/Local_Authority_Districts_(December_2015)_Boundaries.shp')
```

```python
LA_level_dict = {'LA_code' : [] , 'items_per_capita' : [] , 'cost_per_capita' : []}
for name , group in tqdm(presc_2019_with_wards.groupby('LAD15CD')):
    LA_level_dict['LA_code'].append(name)
    LA_level_dict['items_per_capita'].append(sum(group['Total_items'])/(sum(group['All Ages'])/12))
    LA_level_dict['cost_per_capita'].append(sum(group['Total_cost'])/(sum(group['All Ages'])/12))
LA_level_df = pd.DataFrame.from_dict(LA_level_dict)
```

```python
LA_level_df.head(n=4)
```

```python
England_LA_df = pd.merge(lad_boundries,LA_level_df,left_on='lad15cd',right_on='LA_code')
```

```python
England_LA_df.sort_values(by='items_per_capita',inplace=True)
```

```python
England_LA_df.head(n=4)
```

```python
England_LA_df_top10 = England_LA_df[-10:]
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20,20)
ax.axis('off')
ax.set_title('Prescription Items per capita', fontdict={'fontsize': '25', 'fontweight' : '500'})
sm = plt.cm.ScalarMappable(cmap='GnBu', norm=plt.Normalize(vmin=England_LA_df['items_per_capita'].min(), vmax=England_LA_df['items_per_capita'].max()))
#empty array for the data range
sm._A = []
#add the colorbar to the figure
cbar = fig.colorbar(sm,shrink=0.3)
England_LA_df.plot(ax=ax,column='items_per_capita', cmap='GnBu', linewidth=0.8)
England_LA_df_top10.apply(lambda x: ax.annotate(s=x.lad15nm, xy=x.geometry.centroid.coords[0], ha='right',fontsize=20),axis=1);
```
